## 3Dircadb dataset prep

```python
from medloader.api import DatasetEngine
from medloader.app import config as config
from medloader.app.logger import init_logger

logger = init_logger()

cfg = config.Configuration().get("datasets")

engine = DatasetEngine(
    logger=logger,
    config=cfg,
)
```

```python
# Read 3Dircadb
# Launch dataset Raw file processing into a volume in .npy format.

import matplotlib.pyplot as plt
import numpy as np
import os 

engine.go(mode="setup_processed")

slice_id = 50

mask = np.load(
    os.path.join(
        "/dataset/ircadb",
        'processed/3Dircadb1.5',
        f'segmentation.npy')
).astype(float)
plt.imshow(mask[:,:, 52])

```

```python
# Process the clipped dataset images.
import matplotlib.pyplot as plt
import numpy as np
import os 

engine.go(mode="setup_clipping")

slice_id = 94

mask = np.load(
    os.path.join(
        "/dataset/ircadb",
        'clipped/3Dircadb1.5',
        f'segmentation_{slice_id}.npy')
).astype(float)
plt.imshow(mask)
```

## Lits dataset prep

```python
from medloader.api import DatasetEngine
from medloader.app import config_lits as config
from medloader.app.logger import init_logger

logger = init_logger()

cfg = config.Configuration().get("datasets")

engine = DatasetEngine(
    logger=logger,
    config=cfg,
)

# Process the clipped dataset images.
import matplotlib.pyplot as plt
import numpy as np
import os 

engine.go(mode="setup_clipping")

slice_id = 94

mask = np.load(
    os.path.join(
        "/dataset/Lits",
        'clipped/scan_5',
        f'segmentation_{slice_id}.npy')
).astype(float)
plt.imshow(mask)
```
