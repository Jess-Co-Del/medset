```python
from medloader.api import DatasetEngine
from medloader.app import config_lits as config
from medloader.app.logger import init_logger
import torch


logger = init_logger()

cfg = config.Configuration().get("datasets")

engine = DatasetEngine(
    logger=logger,
    config=cfg,
).go(mode="config")

batch_size = 8
from torch.utils.data import DataLoader

train_dl = DataLoader(
    dataset=engine[0],
    batch_size=batch_size,
    pin_memory=True,
    shuffle=False,
    #sampler=DistributedSampler(dataset)
)

val_dl = DataLoader(
    dataset=engine[1],
    batch_size=batch_size,
    pin_memory=True,
    shuffle=False,
    #sampler=DistributedSampler(dataset)
)

train_dl
```
